```python
import pandas as pd
from pathlib import Path
import csv
import requests
import json
from api_keys import api_key
```

```python
movie_file = Path("clean_movies_3.csv")

movie_df = pd.read_csv(movie_file)

movie_df['Director'].count()
```

```python
movie_df[['Actors_O', 'Director_O', 'Rotten_Tomato_O', 'Metascore_O']] = ""

movie_df
# ['Actors']
# ['Director']
# ['Ratings'][1]['Rotten Tomatoes']
# ['Metascore']
```

```python
apiKey = api_key



for index, row in movie_df.iterrows():

    movie = row['title']

    base_url = "http://www.omdbapi.com/"
    url = f"{base_url}?t={movie}&apiKey={apiKey}"

    movie_response = requests.get(url)

    movie_json = movie_response.json()

    

    try:
        director = movie_json["Director"]
        movie_df.loc[index, "Director_O"] = director
    except KeyError:
        movie_df.loc[index, "Director_O"] = "NaN"

    try:
        actors = movie_json["Actors"]
        movie_df.loc[index, "Actors_O"] = actors
    except KeyError:
        movie_df.loc[index, "Actors_O"] = "NaN"

    try:
        metascore = movie_json["Metascore"]
        movie_df.loc[index, "Metascore_O"] = metascore
    except (KeyError, IndexError):
        movie_df.loc[index, "Metascore_O"] = "NaN"
    
    try:
        rotten_tomato = movie_json["Ratings"][1]["Value"]
        movie_df.loc[index, "Rotten_Tomato_O"] = rotten_tomato
    except (KeyError, IndexError):
        movie_df.loc[index, "Rotten_Tomato_O"] = "NaN"

    

movie_df.head()
```

```python
selected_columns = movie_df.loc[:, ['title', 'revenue', 'budget', 'profit', 'main_genre', 'Runtime(Mins)', 'Year', 'original_language', 'Director_O', 'Actors_O', 'Metascore_O', 'Rotten_Tomato_O',
                                    'Award Wins', 'Award Nominations']]  # Selecting specific columns

new_df = pd.DataFrame(selected_columns)
new_df.info()
```

```python
merge_df = pd.read_csv("clean_movies_3.csv")

rating_column = merge_df["Rating"]

merged_df = pd.concat([new_df, rating_column], axis=1)

merged_df.info()
```
