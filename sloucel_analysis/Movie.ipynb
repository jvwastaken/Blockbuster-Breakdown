```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
```

```python
df1 = pd.read_csv('Movie_Rawdataset.csv')
df2 = pd.read_csv('oscar.csv')
df2.columns
```

```python
merged_df = pd.merge(df1, df2[['Film','Award', 'Nomination']], left_on='Movie_Title', right_on='Film', how='left')
merged_df.head()
```

```python
print(merged_df.columns)
```

```python
merged_df.drop(columns='Film', inplace=True) 
merged_df.head()
```

```python
renamed_df = merged_df.rename(columns={'Award': 'Oscar Award', 'Nomination': 'Oscar Nomination'})
renamed_df.head()
```

```python
movies_oscars = renamed_df
```

```python
movies_oscars.to_csv('movies_oscars.csv', index=False) 
```
